**This function will add a new column summarizing the measures**

```python
def pm_per_month(row):
    columns = [f'D0{i+1}' if i < 9 else f'D{i+1}' for i in range(31)]
    summation = 0
    j = 0
    for i in columns:
        if row[i] != 0:
            row['total'] += row[i]
            j +=1
    row['total'] /= j
    return row['total']
```
